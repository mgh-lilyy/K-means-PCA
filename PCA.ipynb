## PCA

```python
import numpy as np
import pandas as pd 
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
```

```python
#getdata
data = pd.read_csv("iris.csv")
encoder = LabelEncoder()
data['species'] = encoder.fit_transform(data['species'])

pca =  PCA(n_components=2)
x_reduced = pca.fit_transform(data)

kmeans = KMeans(n_clusters = 3, init = 'k-means++',
                max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x_reduced)

plt.scatter(x_reduced[y_kmeans == 0, 0], x_reduced[y_kmeans == 0, 1], 
            s = 100, c = 'red', label = 'setosa')
plt.scatter(x_reduced[y_kmeans == 1, 0], x_reduced[y_kmeans == 1, 1], 
            s = 100, c = 'orange', label = 'versicolour')
plt.scatter(x_reduced[y_kmeans == 2, 0], x_reduced[y_kmeans == 2, 1],
            s = 100, c = 'blue', label = 'virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100, c = 'black', label = 'centroids')

plt.legend()



```
